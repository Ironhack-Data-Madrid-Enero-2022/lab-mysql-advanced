```python
db=conn.connect(host='localhost',
                     user='root',
                     passwd='admin',
                     database='pubs')
cursor=db.cursor()
```

```python
cursor.execute('show tables')
for x in cursor:
    print (x)
```

```python
cursor.execute(             
               '''create temporary table pubs.royalty
               SELECT  a.au_id AS "AUTHOR ID",  
               tl.price AS "price",   
               tl.title AS "Title",
               tl.price * sa.qty * tl.royalty / 100 * ta.royaltyper / 100 AS "royal_sales"
               FROM authors AS a
               LEFT JOIN titleauthor AS ta ON a.au_id = ta.au_id
               LEFT JOIN titles AS tl ON tl.title_id = ta.title_id
               LEFT JOIN publishers AS pb ON pb.pub_id = tl.pub_id
               LEFT JOIN sales AS sa ON sa.title_id = ta.title_id
               order by 'royal_sales' desc;''')
```

```python
pd.read_sql('''select* from royalty''',db)
```

```python
cursor.execute('''create temporary table pubs.royaltyauthor
                SELECT  a.au_id AS "AUTHOR ID", 
                tl.price AS "price",   
                tl.title AS "Title",
                sum(tl.price * sa.qty * tl.royalty / 100 * ta.royaltyper / 100) AS "royal_sales"
                FROM authors AS a
                LEFT JOIN titleauthor AS ta ON a.au_id = ta.au_id  
                LEFT JOIN titles AS tl ON tl.title_id = ta.title_id
                LEFT JOIN publishers AS pb ON pb.pub_id = tl.pub_id 
                LEFT JOIN sales AS sa ON sa.title_id = ta.title_id
                group by tl.title;''')
```

```python
pd.read_sql('''select* from royaltyauthor''',db)
```

```python
cursor.execute('''create temporary table pubs.royalfroupsauthottitles
                Select * from
                    (SELECT  a.au_id AS "AUTHOR ID", 
                    tl.price AS "price",   
                    tl.title AS "Title",
                    sum(tl.price * sa.qty * tl.royalty / 100 * ta.royaltyper / 100) AS "royal_sales"
                    FROM authors AS a
                    LEFT JOIN titleauthor AS ta ON a.au_id = ta.au_id  
                    LEFT JOIN titles AS tl ON tl.title_id = ta.title_id
                    LEFT JOIN publishers AS pb ON pb.pub_id = tl.pub_id 
                    LEFT JOIN sales AS sa ON sa.title_id = ta.title_id
                    group by a.au_id) as tabla
                group by Title;''')
```

```python
pd.read_sql('''select* from royalfroupsauthottitles''',db)
```
