```python
tabla='''

SELECT a.au_id AS "AUTHOR ID", tl.title AS "TITLE",(tl.price * sa.qty * (tl.royalty / 100) * (ta.royaltyper / 100)) as SALES_ROYALTY
FROM authors AS a
LEFT JOIN titleauthor AS ta 
ON a.au_id = ta.au_id  
LEFT JOIN titles AS tl 
ON tl.title_id = ta.title_id 
LEFT JOIN publishers AS pb 
ON pb.pub_id = tl.pub_id 
LEFT JOIN sales AS sa 
ON sa.title_id = ta.title_id
ORDER BY SALES_ROYALTY DESC
            
'''

cursor.execute(tabla)
```
