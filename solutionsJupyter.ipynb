```python
from sqlalchemy import create_engine
import pandas as pd
```

```python
str_conn='mysql+pymysql://root:*******.@localhost:3306/publications'

motor=create_engine(str_conn) 
```

```python
pd.read_sql('''select DISTINCTROW authors.au_lname as 'LAST NAME', authors.au_fname as 'FIRST NAME', titles.title as 'TITLE', titles.title_id as 'IDTITULO',
(titles.price * sales.qty * titles.royalty/100*t.royaltyper/ 100) as ROYALTY2
from authors
left join titleauthor as t 
on authors.au_id = t.au_id
left join titles
on t.title_id = titles.title_id
left join sales
on titles.title_id = sales.title_id
group by (qty)''', motor)
```
